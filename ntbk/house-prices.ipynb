# Zbadanie rozkładu ceny

```python
def distplot(data):
    plt.hist(data, bins=50, alpha=0.6, density=True)
    xmin, xmax = plt.xlim()
    mu, std = st.norm.fit(data)
    x = np.linspace(xmin, xmax, 100)
    p = st.johnsonsu.pdf(x, mu, std)
    plt.plot(x, p, 'k', linewidth=2)

    
    pass

       

y = train["SalePrice"]
mean = y.mean()
std = y.std()
plt.figure(1); plt.title("Johnson SU")
distplot(y)
plt.figure(2); plt.title("Normal")
distplot(y)
plt.figure(3); plt.title("Log Norm")
distplot(y)
```
