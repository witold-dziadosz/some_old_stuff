```python
import pandas as pd
```

```python
dax = pd.read_csv("./data/dax-data.csv")
```

```python
dax
```

```python
dax = dax.drop(dax.columns[0], axis=1)
```

```python
def get_lags(df, c, n):
    for i in range(1, n+1):
        df[c + "_lag_" + str(i)] = df[c].shift(i)
```

```python
def get_daytype(r):
    if r.high >= r.high_lag_1 and r.low >= r.low_lag_1:
        ret = "up"
    elif r.high >= r.high_lag_1 and r.low <= r.low_lag_1:
        ret = "outside"
    elif r.high <= r.high_lag_1 and r.low >= r.low_lag_1:
        ret =  "inside"
    elif r.high <= r.high_lag_1 and r.low <= r.low_lag_1:
        ret = "down"
    else:
        ret = "wtf"
    return ret
```

```python
dax.info()
```

```python
# Daytype
cp = dax[["date", "ibdir", "high", "low", "daydir"]].copy()
get_lags(cp, "high", 1)
get_lags(cp, "low", 1)
cp["daytype"] = cp.apply(get_daytype, axis=1)
get_lags(cp, "daytype", 27)
cp = cp.dropna()
cp.info()
```

```python
data = cp[["date", "ibdir", "daytype_lag_7", "daytype_lag_27", "daydir"]].copy()
```
